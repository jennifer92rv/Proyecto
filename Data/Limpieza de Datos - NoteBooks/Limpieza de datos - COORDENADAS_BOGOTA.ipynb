```python
#Se importan las librerias que se van a utilizar
import pandas as pd
import numpy as np
import glob
import os
```

```python
## Se invoca el documento de excel 
df=pd.read_excel('../data/Coordenadas_localidades.xlsx') 


#Se imprimen los datos del DataFrame para validar que se hayan cargado correctamente

df
```

```python
#Se renombran las columnas Nombre gp y CODIGO
df.rename(columns={'gp':'COORDENADAS','CODIGO': 'CODIGO_LOCALIDAD'},inplace=True)
```

```python

#Se reorganiza el orden de las columnas
df = pd.DataFrame.reindex(df,columns =['CODIGO_LOCALIDAD','LOCALIDAD','LONGITUD','LATITUD','COORDENADAS'])

```

```python
#Se imprimen los datos del DataFrame para validar que se hayan aplicado correctamente los cambios
df
```

```python
#Se exporta el DataFrame en un archivo tipo xlsx con el formato utf-8
df.to_excel('Coordenadas_localidades_Bogota.xlsx',encoding="utf-8", index=False)
```
