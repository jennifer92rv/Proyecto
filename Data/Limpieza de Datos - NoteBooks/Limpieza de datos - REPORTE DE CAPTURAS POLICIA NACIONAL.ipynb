REPORTE CAPTURAS POLICIA A NACIONAL

```python
#Importar librerias requeridas
import pandas as pd
import numpy as np
```

```python
## Se invoca el documento de excel que contiene los datos a trabajar 934372 registros.
df1=pd.read_csv('Documents/Proyecto de grado/Datos/REPORTE CAPTURAS POLICIA A NACIONAL.csv', sep =';') 
df1
```

LIMPIEZA DE DATOS

```python
##1.)Se elimina columna Código DANE
df2=df1.drop(['CODIGO DANE'], axis=1)
df2
```

```python
##2.) Se realiza filtro donde el municipio sea igual a Bogotá
df3=df2[df2.MUNICIPIO=="BOGOTÁ D.C. (CT)"]
df3
```

```python
##3 Se Quita la palabra CT del campo MUNICIPIO
df3.loc[df3['MUNICIPIO'] == 'BOGOTÁ D.C. (CT)', 'MUNICIPIO'] = 'BOGOTÁ D.C'
df3
```

```python
##4.) Se cambia el nombre del campo GRUPO ETARIO por POBLACIÓN
df4 = df3.rename(columns={'GRUPO ETARIO':'POBLACION'})
df4.columns
df4
```

```python
##5.) Se eliminan celdas vacias del campo POBLACIÓN  --- DEJAR CAMPO SIN POBLACIÓN nicolas
df5=df4.dropna(subset=["POBLACION"])
df5
```

```python
##6.) Se cambia el nombre del campo DESCRIPCION CONDUCTA CAPTURA por DELITO
df6 = df5.rename(columns={'DESCRIPCION CONDUCTA CAPTURA':'DELITO'})
df6.columns
df6
```

```python
##7.Se filtran los delitos que se van usar en el proyecto
df7 =  df6.query("DELITO == 'ARTÍCULO 103. HOMICIDIO'\
                  or DELITO == 'ARTÍCULO 111. LESIONES PERSONALES'\
                  or DELITO == 'ARTICULO 116A LESIONES CON AGENTES QUIMICOS ACIDO O SUSTANCIAS SIMILARES'\
                  or DELITO == 'ARTÍCULO 135. HOMICIDIO EN PERSONA PROTEGIDA.'\
                  or DELITO == 'ARTÍCULO 239. HURTO PERSONAS' \
                  or DELITO == 'ARTÍCULO 239. HURTO ENTIDADES COMERCIALES'\
                  or DELITO == 'ARTÍCULO 239. HURTO AUTOMOTORES'\
                  or DELITO == 'ARTÍCULO 239. HURTO RESIDENCIAS'\
                  or DELITO == 'ARTÍCULO 239. HURTO ENTIDADES FINANCIERAS'\
                  or DELITO == 'ARTÍCULO 239. HURTO MOTOCICLETAS'\
                  or DELITO == 'ARTÍCULO 239. HURTO DE BIENES PATRIMONIO CULTURAL DE LA NACIÓN'")
df7
```

```python
##8. Se unifica articulos 165 y 166
df8=df7.replace({"ARTÍCULO 166. DESAPARICIÓN FORZADA (CIRCUNSTANCIAS AGRAVACIÓN)": "ARTÍCULO 165. DESAPARICIÓN FORZADA"})
df8
```

```python
##9.)Se filtran datos desde el año 2018 hast 2022
df8['FECHA'] = pd.to_datetime(df8['FECHA HECHO'])
df8['AÑO'] = df8['FECHA'].dt.year
df8 =  df8.query("AÑO in (2018,2019,2020,2021,2022)")
```

```python
##10.Se unifican delitos
df8.loc[df8['DELITO'] == 'ARTICULO 116A LESIONES CON AGENTES QUIMICOS ACIDO O SUSTANCIAS SIMILARES', 'DELITO'] = 'ARTÍCULO 111. LESIONES PERSONALES'
df8.loc[df8['DELITO'] == 'ARTÍCULO 135. HOMICIDIO EN PERSONA PROTEGIDA.', 'DELITO'] = 'ARTÍCULO 103. HOMICIDIO'
```

```python
##11.Se crea campo de articulo y se asignan los códigos de los articulos
conditionlist = [
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO PERSONAS'),
    (df8['DELITO'] == 'ARTÍCULO 111. LESIONES PERSONALES'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES COMERCIALES'),
    (df8['DELITO'] == 'ARTÍCULO 103. HOMICIDIO'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO AUTOMOTORES'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO RESIDENCIAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES FINANCIERAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO MOTOCICLETAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO DE BIENES PATRIMONIO CULTURAL DE LA NACIÓN')]
choicelist = ['239_HP', 
              '111_LP',
              '239_HEC',
              '103_HO',
              '239_HA',
              '239_HR',
              '239_HEF',
              '239_HM',
              '239_HB']
df8['LLAVE_ARTICULO'] = np.select(conditionlist, choicelist, default='Not Specified')
df8
```

```python
##12.Se crea campo de articulo y se asignan los códigos de los articulos
conditionlist = [
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO PERSONAS'),
    (df8['DELITO'] == 'ARTÍCULO 111. LESIONES PERSONALES'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES COMERCIALES'),
    (df8['DELITO'] == 'ARTÍCULO 103. HOMICIDIO'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO AUTOMOTORES'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO RESIDENCIAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES FINANCIERAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO MOTOCICLETAS'),
    (df8['DELITO'] == 'ARTÍCULO 239. HURTO DE BIENES PATRIMONIO CULTURAL DE LA NACIÓN')]
choicelist = [239, 
              111,
              239,
              103,
              239,
              239,
              239,
              239,
              239]
df8['ARTICULO'] = np.select(conditionlist, choicelist, default='Not Specified')
df8
```

```python
##13. quita la palabra articulo y se deja solo descripción del delito
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO PERSONAS', 'DELITO'] = 'HURTO PERSONAS'
df8.loc[df8['DELITO'] == 'ARTÍCULO 111. LESIONES PERSONALES', 'DELITO'] = 'LESIONES PERSONALES'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES COMERCIALES', 'DELITO'] = 'HURTO ENTIDADES COMERCIALES'
df8.loc[df8['DELITO'] == 'ARTÍCULO 103. HOMICIDIO', 'DELITO'] = 'HOMICIDIO'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO AUTOMOTORES', 'DELITO'] = 'HURTO AUTOMOTORES'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO RESIDENCIAS', 'DELITO'] = 'HURTO RESIDENCIAS'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO ENTIDADES FINANCIERAS', 'DELITO'] = 'HURTO ENTIDADES FINANCIERAS'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO MOTOCICLETAS', 'DELITO'] = 'HURTO MOTOCICLETAS'
df8.loc[df8['DELITO'] == 'ARTÍCULO 239. HURTO DE BIENES PATRIMONIO CULTURAL DE LA NACIÓN', 'DELITO'] = 'HURTO DE BIENES PATRIMONIO CULTURAL DE LA NACIÓN'
```

```python
##14.Se reorganiza las columnas
df8 = pd.DataFrame.reindex(df8,columns = ['LLAVE_ARTICULO','DEPARTAMENTO', 'MUNICIPIO', 'FECHA','AÑO', 'GENERO','ARTICULO','DELITO','POBLACION','CANTIDAD'])
```

```python
##15.Se exporta resultado final
df8.to_excel('reportes_de_capturas.xlsx',encoding="utf-8", index=False)
```
