Limpieza de datos - REPORTE DE RECUPERACION DE VEHICULOS POLICIA NACIONAL

```python
import pandas as pd
import numpy as np
```

```python
## Se invoca el documento de excel que contiene los datos a trabajar 148.078 registros
vh1=pd.read_csv('Documents/Proyecto de grado/Datos/RECUPERACION VEHICULOS POLICIA NACIONAL.csv', sep =';') 
vh1
```

```python
##Se filtra dataframe vh1 por Departamento Cundinamarca y Municipio Bogotá D.C

vh2=vh1[(vh1['DEPARTAMENTO'] == 'CUNDINAMARCA') &  (vh1['MUNICIPIO'] == 'BOGOTÁ D.C. (CT)')] 
vh2
```

```python
##Quitar CT del campo MUNICIPIO
vh2.loc[vh2['MUNICIPIO'] == 'BOGOTÁ D.C. (CT)', 'MUNICIPIO'] = 'BOGOTÁ D.C'
vh2
```

```python
## Continuamos eliminando la columna Codigo Dane
vh3=vh2.drop(['CODIGO DANE'], axis=1)
vh3
```

```python
##8.)Se filtran datos desde el año 2018 hast 2022
vh3['FECHA'] = pd.to_datetime(vh3['FECHA HECHO'])
vh3['AÑO'] = vh3['FECHA'].dt.year
vh3 =  vh3.query("AÑO in (2018,2019,2020,2021,2022)")
vh3
```

```python
##9.Se crea campo de llave articulo y se asignan los códigos de los articulos
conditionlist = [
    (vh3['CLASE BIEN'] == 'MOTOCICLETA')]
choicelist = ['239_HM']
vh3['LLAVE_ARTICULO'] = np.select(conditionlist, choicelist, default='239_HA')
vh3
```

```python
##Se exporta resultado final
vh3.to_excel('reportes_de_vehiculos.xlsx',encoding="utf-8", index=False)
```
